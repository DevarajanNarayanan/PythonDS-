```python
#2) Import the required classification libraries along with pandas, numpy, seaborn etc
#3) Then import the classifiers from them (Randomforest, SVM, NaiveBayes, DecisionTrees)
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
#1) Import the train and test csv.
#Read the data file
dataset_train=pd.read_csv("train.csv")
dataset_test=pd.read_csv("test.csv")
```

```python
#4) After this create a function to encode the labels of the strings given in the dataset
x=dataset_train.iloc[:,2:].values
```

```python
y=dataset_train.iloc[:,1].values
```

```python
x_dataset_test=dataset_test.iloc[:,1:].values
```

```python
#5) You can do the above step using label encoder. With this you are creating some labels
#from train set as test set. The test set we imported is for testing the best classifier
#accuracy once we choose it
from sklearn.preprocessing import LabelEncoder
labelx=LabelEncoder()
y=labelx.fit_transform(y)
```

```python
y
```

```python
#6) Then extract the values from train set by stratifying them and dividing it into 80:20 ratio
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
#Model - KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
class_k=KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
class_k.fit(x_train, y_train)
```

```python
#Predict
y_pred_k = class_k.predict(x_test)
```

```python
#Predict _Test
y_pred_k_test = class_k.predict(x_dataset_test)
```

```python
y_pred_k
```

```python
y_pred_k_test
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_k)
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(x_train,y_train)
```

```python
y_pred_nb = classifier_nb.predict(x_test)
#Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_k)
```

```python
y_pred_nb
```

```python
#SVM - Linear
from sklearn.svm import SVC
class_linear=SVC(kernel='linear')
class_linear.fit(x_train,y_train)
```

```python
y_pred_linear=class_linear.predict(x_test)
```

```python
#Accuracy
accuracy_score(y_test, y_pred_linear)
```

```python
#SVM - RBF
from sklearn.svm import SVC
class_rbf=SVC(kernel='rbf')
class_rbf.fit(x_train,y_train)
```

```python
y_pred_rbf=class_rbf.predict(x_test)
```

```python
#Accuracy
accuracy_score(y_test, y_pred_rbf)
```

```python
#SVM - Sigmoid
from sklearn.svm import SVC
class_sigmoid=SVC(kernel='sigmoid')
class_sigmoid.fit(x_train,y_train)
```

```python
y_pred_sigmoid=class_sigmoid.predict(x_test)
```

```python
#Accuracy
accuracy_score(y_test, y_pred_sigmoid)
```

```python
# # #Entropy - Decision Tree
from sklearn.tree import DecisionTreeClassifier
class_tree=DecisionTreeClassifier(criterion="entropy")
class_tree.fit(x_train,y_train)
```

```python
y_pred_tree=class_tree.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_tree)
```

```python
from sklearn.ensemble import RandomForestClassifier
class_random = RandomForestClassifier(n_estimators=3, criterion="entropy")
class_random.fit(x_train,y_train)
```

```python
y_pred_random=class_random.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_random)
```
